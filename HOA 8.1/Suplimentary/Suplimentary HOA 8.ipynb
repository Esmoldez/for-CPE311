```python
import pandas as pd

data = pd.read_csv('data/earthquakes.csv')

jpn_earthquake = data[
    (data['parsed_place'] == 'Japan') & 
    (data['magType'] == 'mb') & 
    (data['mag'] >= 4.9)
]
jpn_earthquake

#Filters only for japan and magnitude of more than 4.9
```

```python
num_magnitude = (
    data[data['magType'] == 'ml']['mag']
    .pipe(pd.cut, bins=range(11), right=False)
    .value_counts()
    .sort_index()
)

num_magnitude
# .pipe is used to make the flow of the code, readable
```

```python
monthly_aggregations = (
    pd.read_csv('data/faang.csv', index_col='date', parse_dates=True)
    .groupby('ticker')
    .resample('M')
    .agg({
        'open': 'mean',
        'high': 'max',
        'low': 'min',
        'close': 'mean',
        'volume': 'sum'
    })
)

monthly_aggregations

#Computes the total of volume traded 
#Computes the Maximum and the minimum prices of high and low respectively
```

```python
df4 = data.copy()

crosstab_max_magnitude = pd.crosstab(df4['tsunami'], 
                                     df4['magType'], 
                                     values=df4['mag'], 
                                     aggfunc='max')

crosstab_max_magnitude
```
